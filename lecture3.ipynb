```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1.5, 1.3, 50)
plt.plot(x, x, 'b--', x, x ** 2, 'g:', x, x ** 3, 'm^', linewidth=2)
plt.grid('on')
plt.legend(['$x$', '$x^2$', '$x^3$'], loc ='best')
plt.show()
```
